```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"data0901.txt", header=None, sep = '\t',names=['id','weight','day','one','age', 'height', 'womenweight', 'smoke'])
```

```python
data = data.drop(data[data['womenweight'] == 999].index)
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] #
plt.rcParams['axes.unicode_minus'] = False #

plt.figure(figsize=(10, 10))
# plt.figure(1)
ax1 = plt.subplot(221)
plt.scatter(x=data["weight"], y=data["age"], color="r", label="age")
plt.legend(loc="best")
ax2 = plt.subplot(222)
plt.scatter(x=data["weight"], y=data["height"], color="g", label="height")
plt.legend(loc="best")
ax3 = plt.subplot(223)
plt.scatter(x=data["weight"], y=data["womenweight"], color="b", label="womenweight")
plt.legend(loc="best")
ax4 = plt.subplot(224)
plt.scatter(x=data["weight"], y=data["smoke"], color="c", label="smoke")


plt.legend(loc="best")



plt.show()
```
